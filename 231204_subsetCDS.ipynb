## subset regions of GTF file
getting CDS, 3' UTR and 5' UTR regions 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gtf = pd.read_csv("/proj/RNA_lab/Genomic_Indexes/Human/Homo_sapiens.GRCh38.108.chr.gtf", skiprows=5, header=None, sep="\t")
```

```python
gtf
```

### CDS 

```python
cds = gtf[gtf[2] == "CDS"]
```

```python
cds.shape
```

```python
cds_annot = cds[8].str.split(";", expand=True)
cds_annot
```

```python
type(cds_annot)
```

```python
cds = cds.drop(columns=8)
cds_complete = pd.concat(objs=[cds, cds_annot], axis=1)
cds_complete
```

### 3' UTR

```python
# following the same method

utr3 = gtf[gtf[2] == "three_prime_utr"]
utr3
```

```python
utr3.shape
```

```python
utr3_annot = utr3[8].str.split(";", expand=True)
utr3 = utr3.drop(columns = 8)
utr3_complete = pd.concat(objs=[utr3, utr3_annot], axis=1)

utr3_complete
```

### 5' UTR

```python
utr5 = gtf[gtf[2] == "five_prime_utr"]
utr5
```

```python
utr5.shape
```

```python
utr5_annot = utr5[8].str.split(";", expand=True)
utr5 = utr5.drop(columns = 8)
utr5_complete = pd.concat(objs=[utr5, utr5_annot], axis = 1)

utr5_complete
```

### Write the files 

```python
# comment out when done saving 
cds_complete.to_csv("/nas/longleaf/home/askav/rna/autoregulation/g4/CDS.subset.gtf", header=None)
utr5_complete.to_csv("/nas/longleaf/home/askav/rna/autoregulation/g4/UTR3.subset.gtf", header=None)
utr3_complete.to_csv("/nas/longleaf/home/askav/rna/autoregulation/g4/UTR5.subset.gtf", header=None)
```
